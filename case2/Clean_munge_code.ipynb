```python
import pandas as pd
```

```python
data = pd.read_pickle('trust_data.p')
```

```python
def re_map(face_id,rating):
    '''
    takes two series objects and creates a dictionary
    with face id as the key and trustworthy rating as the answer
    '''
    scoring = {}
    #for k,v in zip(face_id[6:],rating[6:]):
    for k,v in zip(face_id[6:-1],rating[6:-1]):
        scoring[k] = v
    return scoring
```

```python
def change_by_map(face_id,scoring):
    '''
    takes faceid and rating and changes rating using scoring dict
    
    '''
    final=[]
    for face in face_id.tolist()[:-1]:
        final.append(scoring[face])
    final.append(0)
    return(final)
```

```python
big_list = []
subjects = data.subject[::13]

for sub in subjects:
    subset_of_data = data[data.subject == sub]
    face_id = subset_of_data.faceID
    rating = subset_of_data.trustworthyRate
    scoring = re_map(face_id,rating)
    big_list += change_by_map(face_id,scoring)
print(big_list)
```

```python
data.trustworthyRate = big_list
```
